    <ul>
      <li><b>Amount of Cement:</b> Quantity in metric tons; critical binding agent in the mixture.</li>
      <li><b>Amount of Slag:</b> Byproduct additive in metric tons used to enhance strength and durability.</li>
      <li><b>Amount of Fly Ash:</b> Additive in metric tons that improves workability and sustainability.</li>
      <li><b>Amount of Water:</b> Quantity in liters; essential for hydration and mixture plasticity.</li>
      <li><b>Amount of Superplasticizer:</b> Additive in metric tons to increase flow without extra water.</li>
      <li><b>Amount of Coarse Aggregate:</b> Gravel or crushed stone in metric tons for structural strength.</li>
      <li><b>Amount of Fine Aggregate:</b> Sand in metric tons to enhance mix uniformity and finish.</li>
      <li><b>Slump Value:</b> Workability indicator in centimeters; higher values mean better fluidity.</li>
      <li><b>Flow Value:</b> Measures fluidity in centimeters; useful for self-consolidating concrete.</li>
      <li><b>28-Day Compressive Strength:</b> Strength in MPa; standard test of final concrete quality.</li>
    </ul>

```python
def train_val(y_train, y_train_pred, y_test, y_pred, model_name):
    
    scores = {model_name + "_train" : {"R2" : r2_score(y_train, y_train_pred),
            "mae" : mean_absolute_error(y_train, y_train_pred),
            "mse" : mean_squared_error(y_train, y_train_pred),
            "rmse": np.sqrt(mean_squared_error(y_train, y_train_pred))},
             
              model_name + "_test" : {"R2" : r2_score(y_test, y_pred),
            "mae" : mean_absolute_error(y_test, y_pred),
            "mse" : mean_squared_error(y_test, y_pred),
            "rmse": np.sqrt(mean_squared_error(y_test, y_pred))}}
    return pd.DataFrame(scores)
```

```python
def evaluate_model(name, model, X_test, y_test):
    y_pred = model.predict(X_test)
    return {
        'Model': name,
        'R2 Score': r2_score(y_test, y_pred),
        'RMSE': np.sqrt(mean_squared_error(y_test, y_pred)),
        'MAE': mean_absolute_error(y_test, y_pred)
    }

# Collect results
results = []

# Linear
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
results.append(evaluate_model("Linear Regression", lr_model, X_test, y_test))

# Polynomial Regression (we already trained them here)
for d, model in zip([2, 3, 4], poly_models):
    model.fit(X_train, y_train)
    results.append(evaluate_model(f"Polynomial Regression (deg={d})", model, X_test, y_test))

# Tuned models
results.append(evaluate_model("Lasso (Tuned)", best_lasso, X_test, y_test))
results.append(evaluate_model("Ridge (Tuned)", best_ridge, X_test, y_test))
results.append(evaluate_model("ElasticNet (Tuned)", best_elastic, X_test, y_test))
```

```python

# Define parameter grids
ridge_params = {'alpha': [0.01, 0.1, 1, 10, 100]}
lasso_params = {'alpha': [0.01, 0.1, 1, 10, 100]}
elastic_params = {
    'alpha': [0.01, 0.1, 1, 10],
    'l1_ratio': [0.1, 0.5, 0.9]
}
sgd_params = {
    'alpha': [0.0001, 0.001, 0.01, 0.1],
    'learning_rate': ['constant', 'optimal', 'invscaling', 'adaptive']
}
poly_params = {
    'poly__degree': [2, 3],
    'reg__alpha': [0.01, 0.1, 1, 10]
}

# Dictionary to store R² scores
r2_scores = {}

# Ridge Regression
ridge_grid = GridSearchCV(Ridge(), ridge_params, cv=5, scoring='r2', n_jobs=-1)
ridge_grid.fit(X_train, y_train)
best_ridge = ridge_grid.best_estimator_
print(f"Best Ridge Alpha: {ridge_grid.best_params_['alpha']}")

# Lasso Regression
lasso_grid = GridSearchCV(Lasso(), lasso_params, cv=5, scoring='r2', n_jobs=-1)
lasso_grid.fit(X_train, y_train)
best_lasso = lasso_grid.best_estimator_
print(f"Best Lasso Alpha: {lasso_grid.best_params_['alpha']}")

# ElasticNet Regression
elastic_grid = GridSearchCV(ElasticNet(), elastic_params, cv=5, scoring='r2', n_jobs=-1)
elastic_grid.fit(X_train, y_train)
best_elastic = elastic_grid.best_estimator_
print(f"Best ElasticNet Params: {elastic_grid.best_params_}")

# SGD Regression
sgd_grid = GridSearchCV(SGDRegressor(max_iter=1000, tol=1e-3), sgd_params, cv=5, scoring='r2', n_jobs=-1)
sgd_grid.fit(X_train, y_train)
best_sgd = sgd_grid.best_estimator_
print(f"Best SGD Params: {sgd_grid.best_params_}")

# Polynomial Regression (Ridge with PolynomialFeatures)
poly_pipeline = Pipeline([
    ('poly', PolynomialFeatures()),
    ('reg', Ridge())
])
poly_grid = GridSearchCV(poly_pipeline, poly_params, cv=5, scoring='r2', n_jobs=-1)
poly_grid.fit(X_train, y_train)
best_poly = poly_grid.best_estimator_
print(f"Best Polynomial Params: {poly_grid.best_params_}")

# Evaluate Best Models
r2_scores["Best Ridge"] = r2_score(y_test, best_ridge.predict(X_test))
r2_scores["Best Lasso"] = r2_score(y_test, best_lasso.predict(X_test))
r2_scores["Best ElasticNet"] = r2_score(y_test, best_elastic.predict(X_test))
r2_scores["Best SGD"] = r2_score(y_test, best_sgd.predict(X_test))
r2_scores["Best Polynomial"] = r2_score(y_test, best_poly.predict(X_test))

# Print Results
for model, score in r2_scores.items():
    print(f"{model}: R² = {score:.4f}")

best_model = max(r2_scores, key=r2_scores.get)
print(f"\nBest model after tuning: {best_model} with R² = {r2_scores[best_model]:.4f}")
```
