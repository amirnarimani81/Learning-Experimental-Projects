  <p><b>1️⃣ What is Random Forest?</b></p>
  <ul>
    <li>Random Forest is an <b>ensemble learning method</b> that combines multiple <b>decision trees</b> to improve accuracy and reduce overfitting.</li>
    <li>Based on <b>Bagging (Bootstrap Aggregating)</b>, where multiple models are trained on random subsets of data, and their results are combined.</li>
    <li>Works well for <b>classification and regression tasks</b>.</li>
  </ul>

  <p><b>📌 Key Idea:</b> Instead of relying on one decision tree, Random Forest builds many and aggregates their results for more stable predictions.</p>

  <hr>

  <p><b>2️⃣ How Random Forest Works:</b></p>
  <ul>
    <li><b>Bootstrap Aggregating (Bagging):</b> Train each tree on a random sample (with replacement).</li>
    <li><b>Random Feature Selection:</b> At each split, only a subset of features is considered to decorrelate trees.</li>
    <li><b>Prediction Aggregation:</b> Majority vote (classification) or averaging (regression).</li>
  </ul>

  <p><b>📍 Simple Example:</b> Predict if a patient has a disease based on Age, BP, and Cholesterol.</p>

  <hr>

  <p><b>3️⃣ Mathematical Foundation:</b></p>
  <p><b>Bootstrapping:</b> For a dataset D of size N:</p>
  <p style="font-family: monospace;">D* = {Xᵢ, Yᵢ} where Xᵢ, Yᵢ ∼ D (sampled with replacement)</p>

  <p><b>Splitting Metrics:</b></p>
  <ul>
    <li><b>Gini:</b> G = 1 - Σpᵢ²</li>
    <li><b>Entropy:</b> H = -Σpᵢ log₂pᵢ</li>
    <li><b>MSE:</b> MSE = (Σ(y - ŷ)²)/n</li>
  </ul>

  <hr>

  <p><b>4️⃣ Worked Example:</b></p>
  <p><b>Classification:</b> If 3 trees vote [Yes, Yes, No] → final = Yes (majority vote)</p>
  <p><b>Regression:</b> If 3 trees predict [200K, 250K, 300K] → final = (200+250+300)/3 = 250K</p>

  <hr>

  <p><b>5️⃣ Evaluation Metrics:</b> Accuracy, Precision, Recall, F1-score (Classification), MSE, MAE, R² (Regression)</p>

  <h4>✅ Classification Metrics</h4>
  <p><b>Accuracy:</b> Accuracy = (TP+TN)/(TP+TN+FP+FN)</p>
  <p><b>Precision:</b> Precision = TP/(TP+FP)</p>
  <p><b>Recall:</b> Recall = TP/(TP+FN)</p>
  <p><b>F1 Score:</b> F1 = 2 × (Precision × Recall)/(Precision + Recall)</p>

  <h4>📈 Regression Metrics</h4>
  <p><b>MAE:</b> MAE = (1/n) Σ|Yᵢ - Ŷᵢ|</p>
  <p><b>MSE:</b> MSE = (1/n) Σ(Yᵢ - Ŷᵢ)²</p>
  <p><b>R² Score:</b> R² = 1 - (Σ(Yᵢ - Ŷᵢ)² / Σ(Yᵢ - Ȳ)²)</p>

  <p><b>Step 1: Compute Gini Impurity Before Splitting</b></p>
  <p>Formula: \( Gini = 1 - \sum p_i^2 \)</p>

  <p>Weighted Gini = 0.444</p>

  <p><b>Step 3: Compute Gini for "Call Duration < 35"</b></p>
  <p>Weighted Gini = 0.375 → Lower → So we choose Call Duration as first split ✅</p>

  <p>
    <strong>📍 1. Gini Index – Credit Risk Assessment (Banking)</strong><br>
    Formula: Gini = 1 - ∑ p<sub>i</sub>²<br>
    Dataset Example: Customer income vs loan default.<br>
    Compute Gini for Income feature → best split → separates high-risk vs low-risk customers.<br>
    <em>Impact:</em> Improved loan approval accuracy, reduced defaults.

  <p>
    <strong>📍 2. Entropy – Customer Segmentation (Retail)</strong><br>
    Formula: Entropy = -∑ p<sub>i</sub> log₂ p<sub>i</sub><br>
    Dataset Example: Purchases vs segment A/B.<br>
    Perfect split using “Buys Electronics” → target marketing more effectively.<br>
    <em>Impact:</em> Higher engagement, optimized campaigns.

    MSE computation → better pricing strategy.<br>
    <em>Impact:</em> Optimized property pricing, maximized revenue.
  </p>

    <strong>📍 4. Out-of-Bag (OOB) Error – Stock Price Prediction</strong><br>
    Formula: OOB Error = ∑(y<sub>i</sub> - ŷ<sub>i</sub>)² / n<br>
    Dataset Example: Historical vs predicted stock prices.<br>
    Low OOB → model generalizes well.<br>
    <em>Impact:</em> Reliable stock forecasting, reduced financial risk.
  </p>

```python
# =================================
# 🔥 STEP 1: Import Required Libraries
# =================================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
import seaborn as sns
from sklearn.model_selection import train_test_split, TimeSeriesSplit
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler, LabelEncoder
from datetime import datetime

# =================================
# 🔥 STEP 2: Load Kaggle Dataset
# =================================
# Dataset: Walmart Sales Forecasting (Kaggle)
df = pd.read_csv('/kaggle/input/walmart-sales-forecast/train.csv')  # Use appropriate dataset path
df.head()
```

```python
# Display dataset information
print(df.info())

# Display dataset information
print(df.head())
```

```python
# =================================
# 🔥 STEP 3: Exploratory Data Analysis (EDA)
# =================================
# Check for missing values
print("\n✅ Checking for Missing Values")
print(df.isnull().sum())

# Convert 'date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Extract time-based features
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
df['day'] = df['Date'].dt.day
df['day_of_week'] = df['Date'].dt.dayofweek

# Drop original 'date' column
df.drop(columns=['Date'], inplace=True)
df.head()
```

```python
# =================================
# 🔥 STEP 4: Convert Sales into Categories (Low, Medium, High)
# =================================
# Define demand levels using quantiles
df['demand_category'] = pd.qcut(df['Weekly_Sales'], q=3, labels=['Low', 'Medium', 'High'])

# Encode categorical labels
label_encoder = LabelEncoder()
df['demand_category'] = label_encoder.fit_transform(df['demand_category'])

# Drop original sales column
df.drop(columns=['Weekly_Sales'], inplace=True)
```

```python
df.head()
```

```python
# =================================
# 🔥 STEP 5: Feature Engineering
# =================================
# Creating lag features (Previous week & month sales)
df['lag_7'] = df['demand_category'].shift(7)
df['lag_30'] = df['demand_category'].shift(30)

# Creating rolling mean features
df['rolling_mean_7'] = df['demand_category'].rolling(window=7).mean()
df['rolling_mean_30'] = df['demand_category'].rolling(window=30).mean()

# Fill missing values after shift/rolling
df.fillna(0, inplace=True)
```

```python
# =================================
# 🔥 STEP 6: Train-Test Split
# =================================
X = df.drop(columns=['demand_category'])  # Features
y = df['demand_category']  # Target variable (Demand Level)

from sklearn.preprocessing import label_binarize

# Ensure y_test is correctly binarized
y_bin = label_binarize(y, classes=np.unique(y))  

# Train-Test Split (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=False)
```

```python
rf = RandomForestClassifier(n_estimators = 100, max_depth=8, min_samples_split=2, min_samples_leaf=2,random_state=42)

rf.fit(X_train, y_train)

# Parameters for Hyperparameter Tuning
# param_grid = {
#     'n_estimators': [100, 200, 300],
#     'max_depth': [10, 20],
#     'min_samples_split': [2, 5, 10],
#     'min_samples_leaf': [1, 2, 4],
# }

```

```python
# =================================
# 🔥 STEP 8: Model Evaluation
# =================================
y_pred = rf.predict(X_test)

# Calculate metrics
accuracy = accuracy_score(y_test, y_pred)
print("\n✅ Model Performance Metrics:")
print(f"Accuracy: {accuracy}")

# Classification Report
print("\n✅ Classification Report:")
print(classification_report(y_test, y_pred, target_names=label_encoder.classes_))

# Confusion Matrix
plt.figure(figsize=(6, 4))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap="Blues", xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.title("Confusion Matrix")
plt.show()
```

```python
# =================================
# 🔥 STEP 9: Feature Importance Analysis
# =================================
feature_importances = pd.Series(rf.feature_importances_, index=X.columns).sort_values(ascending=False)

# Plot feature importance
plt.figure(figsize=(10, 6))
sns.barplot(x=feature_importances.values[:10], y=feature_importances.index[:10], palette="viridis")
plt.title("Top 10 Important Features")
plt.show()
```

```python
# =================================
# 🔥 STEP 1: Import Required Libraries
# =================================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV, TimeSeriesSplit
from sklearn.ensemble import RandomForestRegressor, ExtraTreesRegressor, GradientBoostingRegressor, StackingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
from datetime import datetime

# =================================
# 🔥 STEP 2: Load Kaggle Dataset (Historical Stock Trading Data)
# =================================
df = pd.read_csv("/kaggle/input/apple-inc-historical-stock/apple_stock.csv")  # Use appropriate dataset path

# Display dataset information
df.info()
df.head()
```

```python
# =================================
# 🔥 STEP 3: Exploratory Data Analysis (EDA)
# =================================
# Check for missing values
print("\n✅ Checking for Missing Values")
print(df.isnull().sum())

# Convert 'date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Extract time-based features
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
df['day'] = df['Date'].dt.day
df['day_of_week'] = df['Date'].dt.dayofweek

# Drop original 'date' column
df.drop(columns=['Date'], inplace=True)
```

```python
# =================================
# 🔥 STEP 4: Feature Engineering
# =================================
# Creating lag features (Previous day & week closing prices)
df['lag_1'] = df['Close'].shift(1)
df['lag_7'] = df['Close'].shift(7)

# Creating rolling mean features
df['rolling_mean_7'] = df['Close'].rolling(window=7).mean()
df['rolling_mean_30'] = df['Close'].rolling(window=30).mean()

# Fill missing values after shift/rolling
df.fillna(method='bfill', inplace=True)

# Define features and target
X = df.drop(columns=['Close'])  # Features
y = df['Close']  # Target variable (Stock Price)
```

```python
# =================================
# 🔥 STEP 5: Train-Test Split
# =================================
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=False)

# =================================
# 🔥 STEP 6: Feature Scaling
# =================================
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# =================================
# 🔥 STEP 7: Hyperparameter Tuning with GridSearchCV
# =================================
# param_grid = {
#     'n_estimators': [100, 200, 300],
#     'max_depth': [10, 20, None],
#     'min_samples_split': [2, 5, 10],
#     'min_samples_leaf': [1, 2, 4],
# }

rf = RandomForestRegressor(n_estimators= 100, max_depth=10, min_samples_split=2,random_state=42)
rf.fit(X_train_scaled, y_train)
```

```python
# =================================
# 🔥 STEP 8: Model Evaluation
# =================================
y_pred = rf.predict(X_test_scaled)

# Calculate metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("\n✅ Model Performance Metrics:")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"R2 Score: {r2}")
```

```python
# =================================
# 🔥 STEP 9: Feature Importance Analysis
# =================================
feature_importances = pd.Series(rf.feature_importances_, index=X.columns).sort_values(ascending=False)

# Plot feature importance
plt.figure(figsize=(10, 6))
sns.barplot(x=feature_importances.values[:10], y=feature_importances.index[:10], palette="viridis")
plt.title("Top 10 Important Features")
plt.show()
```
