 <section class="card">
      <h2>✨ Project Overview</h2>
      <p class="small">A Python 3.x scraper that fetches hotel name, price, location, rating, review count and booking link for a Booking.com search page and stores results in a CSV file. Designed for reliability with rotating headers and randomized delays to mimic human browsing.</p>
    </section>

  <section class="card">
      <h2>🔧 Technologies & Libraries</h2>
      <ul>
        <li>Python 3.x</li>
        <li><span class="inline-code">beautifulsoup4</span> — HTML parsing</li>
        <li><span class="inline-code">requests</span> — HTTP requests</li>
        <li><span class="inline-code">lxml</span> — fast parser</li>
        <li><span class="inline-code">pandas</span> (optional) — data validation / preview</li>
      </ul>
    </section>
