```python
# -------------------------------
#  yf.download()
# -------------------------------
data_download = yf.download("TSLA", start="2023-01-01", end="2025-11-17", interval="1wk")
print("=== yf.download() ===")
print(data_download.head())  
```

   Question 2: Use Webscraping to Extract Tesla Revenue Data

</center></p></h1>

```python
URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm'

response = requests.get(URL)
```

```python

# --------------------------
#  response.text
# --------------------------
soup_text = BeautifulSoup(response.text, "html.parser")

print("=== با response.text ===")
print(soup_text.prettify()[:300])  

```

```python

# --------------------------
#  response.content
# --------------------------
soup_content = BeautifulSoup(response.content, "html.parser")
print("\n=== با response.content ===")
print(soup_content.prettify()[:300])  # نمایش 300 کاراکتر اول
```

```python
tables = soup_text.find_all("table")
tables
```

```python
rows = soup_text.find_all('tr')
rows
```

```python
data = []
for row in rows[1:]:
    cols = row.find_all('td')
    cols_text = [col.text.strip() for col in cols]
    data.append(cols_text)
```

```python

cleaned_data = []
for row in data:
    row = row[:2]
    while len(row) < 2:
        row.append(None)
    cleaned_data.append(row)
```

```python
tesla_revenue= pd.DataFrame(cleaned_data, columns=['Date', 'Revenue'])
tesla_revenue['Revenue'] = tesla_revenue['Revenue'].replace(r'[\$,]', '', regex=True)
tesla_revenue.head()
```

  Question 4: Use Webscraping to Extract GME Revenue Data

</center></p></h1>

```python
URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/stock.html'

response1 = requests.get(URL)
```

```python

soup_text1 = BeautifulSoup(response1.text, "html.parser")

print("=== با response.text ===")
print(soup_text1.prettify()[:300])  
```

```python
tables = soup_text1.find_all('table')
revenue_table = tables[1]

revenue_table
```

```python
rows1 = revenue_table.find_all('tr')
rows1
```

```python
data1 = []
for row in rows1[1:]:  # رد header
    cols = row.find_all('td')
    cols_text = [col.text.strip() for col in cols]
    data1.append(cols_text)
```

```python
cleaned_data1 = []
for row in data1:
    row = row[:2]
    while len(row) < 2:
        row.append(None)
    cleaned_data1.append(row)
```

```python
gme_revenue = pd.DataFrame(cleaned_data1, columns=["Date", "Revenue"])
gme_revenue['Revenue'] = gme_revenue['Revenue'].replace(r'[\$,]', '', regex=True)
gme_revenue.head()
```

<h1><p><center style="background: linear-gradient(to right,white,green);color: transparent;font-family: Gabriola;color: navy;font-size: 80%;text-align: center;border-radius: 10px 70px">
  
   Question 5: Plot Tesla Stock Graph


</center></p></h1>

```python
def make_graph(stock_data, revenue_data, stock):
    fig = make_subplots(
        rows=2, cols=1,
        shared_xaxes=True,
        subplot_titles=("Historical Share Price", "Historical Revenue"),
        vertical_spacing=0.3
    )

    # Fixed date typo
    stock_data_specific = stock_data[stock_data.Date <= '2021-06-14']
    revenue_data_specific = revenue_data[revenue_data.Date <= '2021-04-30']

    fig.add_trace(
        go.Scatter(
            x=pd.to_datetime(stock_data_specific.Date),
            y=stock_data_specific.Close.astype(float),
            name="Share Price"
        ),
        row=1, col=1
    )

    fig.add_trace(
        go.Scatter(
            x=pd.to_datetime(revenue_data_specific.Date),
            y=revenue_data_specific.Revenue.astype(float),
            name="Revenue"
        ),
        row=2, col=1
    )

    fig.update_xaxes(title_text="Date", row=1, col=1)
    fig.update_xaxes(title_text="Date", row=2, col=1)
    fig.update_yaxes(title_text="Price ($US)", row=1, col=1)
    fig.update_yaxes(title_text="Revenue ($US Millions)", row=2, col=1)

    fig.update_layout(
        showlegend=False,
        height=900,
        title=stock,
        xaxis_rangeslider_visible=True
    )

    fig.show()
```

```python
tesla_revenue["Date"] = tesla_revenue["Date"].str.replace(r"[^\d-]", "", regex=True)
tesla_revenue["Date"] = pd.to_datetime(tesla_revenue["Date"], errors="coerce")
```

```python
make_graph(data_ticker , tesla_revenue, 'Tesla')
```

```python
make_graph(gme_data, gme_revenue, 'GameStop')
```
